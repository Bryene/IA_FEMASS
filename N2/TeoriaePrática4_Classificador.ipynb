# 2.1 Validação Cruzada:

Esses exemplos ilustram como utilizar a validação cruzada e a matriz de confusão para uma avaliação detalhada e robusta de um classificador binário. A validação cruzada é uma técnica fundamental na avaliação de modelos de aprendizado de máquina, especialmente em cenários de classificação binária. Ela desempenha um papel crucial ao proporcionar uma estimativa confiável do desempenho do modelo. A essência da validação cruzada reside na divisão dos dados em várias dobras ou partições. Isso permite que avaliemos o desempenho do modelo em várias configurações de treinamento e teste. Em cada rodada de validação cruzada, uma parte dos dados é reservada para treinamento e a outra para teste, e esse processo é repetido várias vezes. A média dos resultados obtidos em cada rodada fornece uma avaliação mais robusta e geral do modelo, ajudando a mitigar o impacto de uma única divisão de dados que poderia ser tendenciosa. Por outro lado, a matriz de confusão é uma ferramenta valiosa que oferece uma análise detalhada das previsões feitas pelo modelo. Ela divide as previsões em quatro categorias: verdadeiros positivos, falsos positivos, falsos negativos e verdadeiros negativos. Os verdadeiros positivos representam os casos em que o modelo acertou ao classificar positivamente, os falsos positivos são erros em que o modelo classificou erroneamente como positivo, os falsos negativos são erros em que o modelo classificou erroneamente como negativo e os verdadeiros negativos são os casos corretamente classificados como negativos.




```python
from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

Nesse trecho, é utilizado a função cross_val_score para realizar a validação cruzada do classificador. Definido o parâmetro cv como 3 para indicar que queremos dividir os dados em 3 dobras para validação cruzada. A métrica de "acurácia" é utilizada para avaliar o desempenho do classificador. O resultado é uma lista de pontuações de acurácia para cada dobra.

```python
"""
Este código ilustra a realização da validação cruzada manualmente usando StratifiedKFold,
uma técnica que divide o conjunto de treinamento em folds estratificados.
A validação cruzada é uma abordagem fundamental para avaliar o desempenho do classificador em várias divisões dos dados,
a fim de obter uma estimativa mais robusta.

Este processo fornece uma avaliação do desempenho do classificador
em várias divisões dos dados, permitindo uma análise mais confiável do modelo.
"""
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
# Criação dos folds estratificados usando StratifiedKFold
skfolds = StratifiedKFold(n_splits=3)

# Loop para realizar a validação cruzada manualmente
for train_index, test_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_5[test_index]

    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

Neste trecho de código, realizamos a validação cruzada manual, aproveitando a classe `StratifiedKFold` para criar folds estratificados. Começamos criando uma instância de `StratifiedKFold` com 3 dobras. Em seguida, usamos um loop para iterar sobre essas dobras de treinamento e teste. A cada iteração, clonamos o classificador original (`clone_clf`) para obter uma nova instância, garantindo que cada fold tenha um modelo independente. Depois, selecionamos os dados de treinamento e teste que correspondem às dobras atuais. O classificador é treinado com os dados de treinamento do fold e, em seguida, fazemos previsões nos dados de teste. Calculamos a taxa de acertos comparando essas previsões com os rótulos verdadeiros dos dados de teste e, por fim, imprimimos o resultado. Esse processo nos permite avaliar o desempenho do modelo em cada uma das dobras da validação cruzada de forma manual, fornecendo métricas de desempenho para cada iteração.

```python
from sklearn.dummy import DummyClassifier

dummy_clf = DummyClassifier()
# Treinamento do classificador DummyClassifier
dummy_clf.fit(X_train, y_train_5)
# Realização de uma previsão usando o classificador DummyClassifier
print(any(dummy_clf.predict(X_train)))
```

```python
# Avaliação do classificador DummyClassifier usando validação cruzada
cross_val_score(dummy_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

Nesse trecho de código, criamos um classificador DummyClassifier, treinamos o classificador e realizamos uma previsão usando o método predict(). Em seguida, usamos a função cross_val_score para realizar a validação cruzada do classificador DummyClassifier, obtendo as pontuações de acurácia para cada dobra.

# Limiar de decisão:
O limiar de decisão é um valor que determina o ponto de separação entre as classes positiva e negativa em um classificador binário. É usado para tomar a decisão de classificação com base nos valores de pontuação ou probabilidade produzidos pelo modelo. Alterar o limiar de decisão pode ter um impacto significativo na precisão e revocação do classificador. Aumentar ou diminuir o limiar pode levar a um equilíbrio entre essas métricas, permitindo que os usuários ajustem o modelo para atender às necessidades específicas do problema, priorizando a minimização de falsos positivos ou a identificação de casos verdadeiramente positivos, conforme necessário. Portanto, o limiar de decisão é uma ferramenta poderosa para personalizar o desempenho do classificador de acordo com os requisitos do cenário de aplicação.



```python
from sklearn.metrics import precision_score, recall_score

# Calcula a precisão entre os rótulos verdadeiros (y_train_5) e as previsões (y_train_pred)
precision_score(y_train_5, y_train_pred)  # == 3530 / (687 + 3530)
```

```python
# Calcula a precisão manualmente: TP / (FP + TP)
cm[1, 1] / (cm[0, 1] + cm[1, 1])
```

```python
# Calcula a revocação entre os rótulos verdadeiros (y_train_5) e as previsões (y_train_pred)
recall_score(y_train_5, y_train_pred)  # == 3530 / (1891 + 3530)
```

```python
# Calcula a revocação manualmente: TP / (FN + TP)
cm[1, 1] / (cm[1, 0] + cm[1, 1])
```

# Compensação da Precisão/Revocação:
A precisão e a revocação são métricas cruciais para avaliar o desempenho de um classificador binário, sendo particularmente valiosas quando há uma disparidade significativa entre as classes. Elas fornecem informações vitais sobre a qualidade das classificações do modelo. A precisão mensura a fração de exemplos positivos que foram corretamente identificados, em relação ao total de exemplos classificados como positivos. Por sua vez, a revocação avalia a proporção de exemplos positivos que foram adequadamente classificados em relação ao número total de exemplos que realmente são positivos.Essas métricas são essenciais para entender não apenas a precisão das previsões, mas também a capacidade do modelo de recuperar corretamente todos os exemplos de uma classe específica. Portanto, elas desempenham um papel crítico na avaliação do desempenho e na adequação de um classificador binário, especialmente quando o desequilíbrio entre as classes é um fator a ser considerado.

```python
# Previsões de todas as instâncias de treinamento usando a função de decisão do classificador SGD
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3,
                             method="decision_function")
```

```python
from sklearn.metrics import precision_recall_curve

# Calcula as precisões, revocações e limiares para diferentes limiares de decisão
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
plt.figure(figsize=(8, 4))  # extra code – it's not needed, just formatting
plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
plt.vlines(threshold, 0, 1.0, "k", "dotted", label="threshold")

# Plota um ponto na curva para o limiar específico
idx = (thresholds >= threshold).argmax()  # first index ≥ threshold
plt.plot(thresholds[idx], precisions[idx], "bo")
plt.plot(thresholds[idx], recalls[idx], "go")
plt.axis([-50000, 50000, 0, 1])
plt.grid()
plt.xlabel("Threshold")
plt.legend(loc="center right")
# save_fig("precision_recall_vs_threshold_plot")

plt.show()
```

Podemos nos perguntar por que a curva de precisão é mais irregular do que a curva de revocação nessa figura. A razão é que a precisão às vezes pode diminuir quando você aumenta o limiar (embora em geral ela aumente). Para entender o porquê, veja  e observe o que acontece quando você começa a partir do limiar central e o move apenas um dígito para a direita: a precisão passa de 4/5 (80%) para 3/4 (75%). Por outro lado, a revocação só diminuirá quando o limiar for aumentado, o que explica por que sua curva parece suave.Outra maneira de selecionar uma boa compensação de precisão/revocação é plotar a precisão diretamente contra a revocação.

```python
import matplotlib.patches as patches

plt.figure(figsize=(6, 5))

# Plota a curva de precisão/revocação
plt.plot(recalls, precisions, linewidth=2, label="Precision/Recall curve")

plt.plot([recalls[idx], recalls[idx]], [0., precisions[idx]], "k:")
plt.plot([0.0, recalls[idx]], [precisions[idx], precisions[idx]], "k:")
plt.plot([recalls[idx]], [precisions[idx]], "ko",
         label="Point at threshold 3,000")
plt.gca().add_patch(patches.FancyArrowPatch(
    (0.79, 0.60), (0.61, 0.78),
    connectionstyle="arc3,rad=.2",
    arrowstyle="Simple, tail_width=1.5, head_width=8, head_length=10",
    color="#444444"))
plt.text(0.56, 0.62, "Higher\nthreshold", color="#333333")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.axis([0, 1, 0, 1])
plt.grid()
plt.legend(loc="lower left")
# save_fig("precision_vs_recall_plot")
plt.show()
```

Aqui é selecionado uma boa compensação de precisão/revocação pois foi plotado a precisão diretamente contra a revocação. É possível ver que a precisão realmente começa a diminuir acentuadamente em torno de 80% de revocação. Provavelmente será selecionado compensação de precisão/revocação antes dessa queda — por exemplo, em torno de 60% de revocação. Então, vamos supor que você almeja 90% de precisão. Procura-se a primeira plotagem (aproximando um pouco) e descobre que precisa utilizar um limiar de cerca de 70 mil. Em vez de chamar o método predict() do classificador, pode-se apenas executar este
código para fazer previsões, como visto nos exemplos abaixo:

```python
idx_for_90_precision = (precisions >= 0.90).argmax()
threshold_for_90_precision = thresholds[idx_for_90_precision]
threshold_for_90_precision
```

```python
# Previsões utilizando o limiar de decisão para obter uma precisão de 90%
y_train_pred_90 = (y_scores >= threshold_for_90_precision)
precision_score(y_train_5, y_train_pred_90)
```

```python
# Revocação obtida usando o limiar de decisão para obter uma precisão de 90%
recall_at_90_precision = recall_score(y_train_5, y_train_pred_90)
recall_at_90_precision
```

# A Curva ROC:
A curva ROC, sigla para "Receiver Operating Characteristic," é uma representação gráfica que demonstra de maneira visual como o desempenho de um classificador binário se altera conforme o limiar de decisão é ajustado. Essa curva exibe a relação entre a taxa de verdadeiros positivos (Recall ou TPR), localizada no eixo vertical (y), e a taxa de falsos positivos (Fallout ou FPR), encontrada no eixo horizontal (x). 

```python
"""
Este código utiliza a função roc_curve da biblioteca scikit-learn para calcular as taxas
de FPR (False Positive Rate) e TPR (True Positive Rate) em vários limiares com base nas
pontuações (y_scores) de um classificador binário.

- y_train_5: Array contendo os rótulos verdadeiros (0 ou 1) do conjunto de treinamento.
- y_scores: Array contendo as pontuações ou probabilidades de pertencimento à classe positiva (classe 5)
produzidas pelo classificador.

O código também identifica o ponto na curva ROC que corresponde a um limiar específico para alcançar
uma precisão de 90%. Em seguida, ele cria um gráfico da curva ROC, destacando a curva real, a curva do classificador
aleatório e o ponto associado ao limiar para 90% de precisão.
"""
from sklearn.metrics import roc_curve

# Calcula as taxas de FPR e TPR em vários limiares com base nas pontuações (y_scores)
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
idx_for_threshold_at_90 = (thresholds <= threshold_for_90_precision).argmax()
tpr_90, fpr_90 = tpr[idx_for_threshold_at_90], fpr[idx_for_threshold_at_90]

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, linewidth=2, label="ROC curve")
plt.plot([0, 1], [0, 1], 'k:', label="Random classifier's ROC curve")
plt.plot([fpr_90], [tpr_90], "ko", label="Threshold for 90% precision")

plt.gca().add_patch(patches.FancyArrowPatch(
    (0.20, 0.89), (0.07, 0.70),
    connectionstyle="arc3,rad=.4",
    arrowstyle="Simple, tail_width=1.5, head_width=8, head_length=10",
    color="#444444"))
plt.text(0.12, 0.71, "Higher\nthreshold", color="#333333")
plt.xlabel('False Positive Rate (Fall-Out)')
plt.ylabel('True Positive Rate (Recall)')
plt.grid()
plt.axis([0, 1, 0, 1])
plt.legend(loc="lower right", fontsize=13)
# save_fig("roc_curve_plot")

plt.show()
```

Quanto maior a revocação (TPR), mais falsos positivos (FPR) o classificador produz. A linha pontilhada representa a curva ROC de um classificador puramente aleatório; um bom classificador fica o mais distante dessa linha possível (em direção ao canto superior esquerdo). Uma maneira de comparar classificadores é medir a área abaixo da curva (AUC, do inglês). Um classificador perfeito terá um ROC AUC igual a 1, enquanto um classificador puramente aleatório terá um ROC AUC igual a 0,5. O Scikit-Learn fornece uma função para calcular o ROC AUC:

```python
from sklearn.metrics import roc_auc_score

# Calcula a área sob a curva ROC (ROC AUC)
roc_auc_score(y_train_5, y_scores)
```

A razão é que usamos um SGDClassifier simples, que é um modelo linear. Tudo o que ele faz é atribuir um peso por classe a cada pixel, e, quando ele vê uma nova imagem, apenas resume as intensidades dos pixels analisados para obter uma pontuação para cada classe. Então, como os 3 e 5 diferem apenas em alguns pixels, esse modelo facilmente os confundirá.

# Classificação Multilabel:
A classificação multilabel é uma tarefa de aprendizado de máquina que permite associar múltiplas classes a uma única instância de dados simultaneamente. Diferentemente da classificação multiclasse, na qual cada instância é atribuída a uma única classe exclusiva, a classificação multilabel oferece a flexibilidade de permitir que uma instância pertença a várias classes ao mesmo tempo.

Existem diversas abordagens para lidar com a classificação multilabel. Duas delas são a abordagem "one-vs-rest" (um-contra-o-resto) e a abordagem "one-vs-one" (um-contra-um). Na abordagem "one-vs-rest" (também chamada de "one-vs-all"), um classificador é treinado para cada classe individualmente. Cada classificador trata as instâncias dessa classe como positivas e todas as outras classes como negativas. Portanto, para um problema com N classes, são treinados N classificadores. Durante a previsão, cada classificador gera uma pontuação de confiança para sua classe correspondente, e a classe com a maior pontuação é considerada a classe prevista. Essa abordagem é adequada para problemas de classificação binária e pode ser estendida para problemas multilabel atribuindo a classe com a pontuação mais alta como a classe prevista e, se necessário, usando um limiar para determinar quais classes são atribuídas à instância. Na abordagem "one-vs-one" (um-contra-um), é treinado um classificador para cada par de classes. Cada classificador considera apenas as instâncias pertencentes a essas duas classes. Assim, para um problema com N classes, são treinados N * (N-1) / 2 classificadores. Durante a previsão, cada classificador vota na classe correspondente ao par de classes para o qual foi treinado, e a classe com o maior número de votos é considerada a classe prevista. Essa abordagem é adequada tanto para problemas de classificação multiclasse quanto para problemas multilabel, permitindo a atribuição de várias classes à instância com base nos votos dos classificadores.

Além dessas abordagens, existem algoritmos específicos projetados para tratar diretamente a classificação multilabel, como o algoritmo Binary Relevance, o algoritmo Label Powerset e o algoritmo Classifier Chains. Esses algoritmos abordam a classificação multilabel como uma combinação de problemas binários, em que cada classe é tratada individualmente em relação ao conjunto de todas as outras classes, proporcionando uma gama de abordagens para resolver problemas complexos que envolvem múltiplas classes associadas a cada instância.
